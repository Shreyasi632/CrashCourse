# Understanding Overfitting and Underfitting

### **Overfitting**
A model overfits when it becomes too adept at recognizing the noise and random fluctuations that make up the training data. As a result, the model may become overly complicated and impractical for use with fresh data. To put it another way, rather than discovering underlying patterns that may be used to analyze new data, the model is effectively "memorizing" the training data. High training accuracy but poor performance on new data are frequent indicators of overfitting.


### **Underfitting**
When a model is too basic and unable to recognize the underlying patterns in the training data, underfitting occurs. This could happen if the model isn't complicated enough or if not enough data was used to train the model. In other words, the model is unable to discover the underlying patterns and is unable to make precise predictions as a result. Poor training accuracy and poor performance on new data are frequent signs of underfitting.

# What are the Bias and Variance?

Bias refers to the difference between the average prediction of a model and the true value of the target variable. In other words, bias measures how much a model is systematically underestimating or overestimating the true value of the target variable. A model with high bias is said to be underfitting the data, and is not capturing the underlying patterns in the data.

Variance, on the other hand, refers to the variability of a model's prediction for a given input. In other words, variance measures how much the predictions of a model differ from each other when trained on different subsets of the data. A model with high variance is said to be overfitting the data, and is capturing the noise and random fluctuations in the data.

Ideally, we want a model that has low bias and low variance, meaning that it is both accurate and stable in its predictions. Achieving this balance is known as the bias-variance tradeoff, and is a key consideration in model selection and evaluation.

To achieve a good bias-variance tradeoff, we may use techniques like regularization to reduce the model's complexity and prevent overfitting. We may also use techniques like cross-validation to assess the model's performance on new data and ensure that it is not overfitting or underfitting. Additionally, we may explore different model architectures or hyperparameters to find the best balance between bias and variance.

## What is Regularization ?
Regularization is a technique used in machine learning to prevent overfitting of models. Overfitting occurs when a model becomes too complex, fitting the training data so well that it fails to generalize well on new, unseen data. Regularization aims to solve this problem by adding a penalty term to the loss function of the model, encouraging it to choose simpler and more generalizable models.

The penalty term in the loss function is based on the model's parameters, and it penalizes large values of these parameters. The idea is to discourage the model from assigning too much importance to any one feature, which could lead to overfitting.

##How regularization works:

Regularization works by adding a penalty term to the loss function of the model, which discourages the model from assigning too much importance to any one feature. The size of the penalty is controlled by a hyperparameter, which is set by the user.

The hyperparameter controls the strength of the regularization, with a larger value resulting in a stronger regularization. If the hyperparameter is set too high, the model may become too simple and underfit the data, failing to capture the complex relationships between the features.

Regularization is usually applied during the training phase of the model. During training, the loss function is minimized by adjusting the model's parameters, taking into account the penalty term added by regularization. The final model is then evaluated on the validation set to ensure that it has good generalization performance.

## Advantages of Regularizations 
Regularization is a widely used technique in machine learning that offers several advantages, including:

- **Prevents overfitting:** The main advantage of regularization is that it helps prevent overfitting of models. Overfitting occurs when a model becomes too complex, fitting the training data so well that it fails to generalize well on new, unseen data. Regularization helps to solve this problem by adding a penalty term to the loss function of the model, encouraging it to choose simpler and more generalizable models.

- **Improves generalization:** By reducing overfitting, regularization improves the generalization performance of the model. The model learns to focus on the most relevant features of the data, while ignoring the noise and irrelevant features.

- **Handles noisy data:** Regularization is particularly useful when dealing with noisy data, where the input data has errors or inconsistencies. It helps to smooth out the noise and reduces the impact of outliers.

- **Reduces model complexity:** Regularization encourages the model to choose simpler and more generalizable models. This can lead to a reduction in model complexity, making it easier to interpret and understand the underlying relationships in the data.

- **Increases model stability:** Regularization can help to increase the stability of the model by reducing the variance in the model's predictions. This can improve the reliability of the model and make it less prone to fluctuations in the data.

# Lasso Regression (L1 regularization)
Lasso Regression is a linear regression method used for feature selection and regularization. It is often referred to as L1 regularization. In Lasso Regression, the loss function of the linear regression model is supplemented by a penalty term proportional to the total absolute value of the coefficients. This penalty term promotes the model to have sparse coefficients, which means that it will probably set a lot of coefficients to zero and so choose only the most crucial attributes for the prediction.

Lasso.png


This technique is different from ridge regression as it uses absolute weight values for normalization. λ is again a tuning parameter and behaves in the same as it does when using ridge regression.

As loss function only considers absolute weights, optimization algorithms penalize higher weight values.

# Advantages of Lasso Regularization

Here are some advantages of Lasso regularization:

- **Feature selection:** Lasso regularization tends to shrink the coefficients of less important features to zero, effectively removing them from the model. This makes it a useful tool for feature selection, especially when dealing with high-dimensional datasets with many irrelevant or redundant features.

- **Improved model interpretability:** Lasso regularization can help simplify the model and make it easier to interpret by reducing the number of non-zerp coefficients. This is particularly useful in domains such as finance or healthcare, where the ability to explain model predictions is important.

- **Better generalization:** Lasso regularization can improve the generalization performance of a model by reducing overfitting. This is because it discourages the model from relying too much on any one feature, which can lead to overfitting on the training data and poor performance on new data.

- **Robustness to collinearity:** Lasso regularization can handle collinear features, which are features that are highly correlated with each other. In contrast, other regularization techniques such as ridge regularization may struggle with collinearity and produce unstable estimates.

- **Automatic parameter selection:** Lasso regularization can automatically select the optimal regularization parameter (lambda) using cross-validation, making it easier to tune the model without needing to manually search for the best value.

Overall, Lasso regularization is a powerful tool that can help improve model performance, simplify the model, and enhance interpretability, particularly in high-dimensional datasets with many irrelevant features.

# Disadvantages of Lasso Regularization

- **Biased coefficients:** The coefficients generated by a LASSO model are biased and hence one of the key disadvantages of LASSO regression. The L1 penalty that is applied to the model causes the coefficients to be artificially shrunk, sometimes all the way to zero. This means that the coefficients from a LASSO model indicate a scaled-down version of the true magnitude of the link between the characteristics and the result.


- **Difficult to estimate standard errors:** Since the coefficient estimates in a LASSO model are biased, it is difficult to estimate accurate standard errors for them. This makes it difficult to do things like run statistical tests on them and build confidence intervals around them.

- **Struggle with correlated features:** When trained on data containing linked characteristics, LASSO models also have the drawback of being extremely unstable. Usually, one characteristic is chosen relatively randomly, and all of the other features that are strongly connected with it are effectively eliminated from the model. This could mislead someone into thinking that only the feature that was kept in the model is significant, whereas in fact, some of the other features might be equally or even more significant.


- **Generally unstable estimates:** Since LASSO models are known to provide estimates that are relatively unstable, training them on slightly different datasets might cause significant changes in those estimates. For instance, you may anticipate seeing different characteristics removed from the model for each dataset if you bootstrap the data several times to produce a number of distinct sample datasets. Even if all of the datasets you are using for training are quite comparable, this can still occur.


- **Introduction of a hyperparameter:** Although this drawback is less significant, LASSO models incorporate a hyperparameter to control the severity of the L1 penalty. As a result, you must perform actions related to hyperparameter tuning that are not necessary when using a regular regression model.


- **Other issues associated with standard regression models:** LASSO regression models are also plagued by some of the same issues that affect standard regression models. Concerns surrounding interactions, outliers, and stringent model assumptions also apply to this family of models.

## Ridge Regression (L2 regularization)
Ridge regularization, also known as L2 regularization, is a method used in statistical modeling and machine learning to avoid overfitting by including a penalty element in the loss function. The penalty term promotes the model to have smaller and more evenly distributed coefficients since it is inversely proportional to the square of the magnitude of the model coefficients. When employing this strategy, we modify an existing loss function by adding the square of the weights, as seen here:

Ridge.png

In this case, normalized weights are presented as squares. There may have been parameters and normalized weights, as you could have seen. A cross-validation dataset must be used to fine-tune the parameter. When you enter =0, the initial loss function you choose is given back as the residual sum of squares. When is very high, loss will disregard the primary loss function, reduce the weight's square, and ultimately set the parameters' values to zero. The parameters are currently learned using a modified loss function. Since the aforementioned function must be minimized, parameters must be as little as practical. Thus, the L2 standard limits the rate of weight growth.



# Advantages of Ridge Regularization
Here are some advantages of using ridge regularization:

- **Reduces overfitting:** A model's overfitting can be decreased with the use of ridge regularization. This is accomplished by introducing a penalty term into the loss function that dissuades the model from giving any one characteristic an excessive amount of weight. The model's generalization may be enhanced as a result.


- **Handles multicollinearity:** When dealing with multicollinearity in data, ridge regularization is especially helpful. When two or more features in a dataset have a significant degree of correlation, multicollinearity results. By lessening the influence of linked features on the model, ridge regularization can help with this.


- **Improves model stability:** The stability of a model can be enhanced through ridge regularization. This is due to the fact that it lessens the effect of data noise and makes the model more resilient to subtle input changes.


- **Allows for the inclusion of all features:** Ridge regularization does not force features to be dropped from the model. Instead, it reduces the impact of each feature on the model, allowing all features to be included in the model.

- **Easy to implement:** Ridge regularization is a simple and easy-to-apply technique. Most machine learning algorithms may readily include this need by simply adding a penalty term to the loss function.


# Disadvantages of Ridge Regularization

While ridge regularization can be very effective, it also has some disadvantages:

- **Bias-variance tradeoff:** Ridge regularization can assist avoid overfitting by including a bias component, but it can also introduce bias into the model by requiring the coefficients to be lower than they might otherwise be. As a result, there may be a trade-off between decreasing variance and boosting bias.


- **Feature selection:** Ridge regularization does not perform feature selection, Ridge regularization does not do feature selection, therefore all features, regardless of how relevant they are to the target variable, are kept in the model. This may result in a model with duplicate features that is harder to interpret.


- **Complexity:** With large datasets or numerous features, the ridge regularization approach can be computationally expensive. This may render it useless for various purposes.


- **Hyperparameter tuning:** To attain the greatest results, the regularization parameter in ridge regression must be modified. This can be a difficult and time-consuming process that necessitates a deep comprehension of the model and data.

## Why GLM ?

The GLM model is a simpler and more understandable model, making it simpler to comprehend the correlations between the input variables and the target variable. GLM models the correlation between the input variables and the goal variable using a linear function and presupposes a particular probability distribution for the target variable. As a result, it works effectively in situations when the link between the input variables and the target variable is clear-cut and straightforward.

There are several advantages of using a Generalized Linear Model (GLM):

1. Flexibility: GLMs can handle a wide range of response variables, including continuous, binary, count, and categorical data.

2. Interpretable coefficients: The link between the response variable and the predictor variables can be understood using the interpretable coefficients that are provided by GLMs.

3. Model diagnostics: GLMs offer a range of diagnostic tools that can be used to evaluate the suitability of the model, including leverage diagnostics, deviance goodness-of-fit tests, and residual plots.

4. Efficient estimation: Maximum likelihood estimation, a computationally effective technique, is used by GLMs to estimate model parameters.

5. Easy to implement: GLMs can be implemented quite quickly using statistical software programs like R or Python.

6. Robustness: GLMs are resistant to outliers in the data, therefore even if the dataset contains extreme values, they can still provide reliable predictions.

Overall, GLMs are a well-liked option in many domains, including biostatistics, epidemiology, and economics, as they are a strong and adaptable modeling framework that can be used to examine a wide range of data sources.

### Identifing predictor significance using OLS regression
Ordinary least squares (OLS) regression is a statistical method of analysis used for multivariate model that estimates the relationship between one or more independent variables and a dependent variable; the method estimates the relationship by minimizing the sum of the squares in the difference between the observed and predicted values of the dependent variable.
